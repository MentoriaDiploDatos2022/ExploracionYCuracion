### 2. Análisis de outliers: 



```python
# Creamos un nuevo df para eliminar las columnas que no queremos analizar
all_labeled_curado_reducido = all_labeled_curado.drop(columns = ['Patient', 'State'])
```

```python
# Hicimos un análisis general simétrico de outliers 
# Tomamos como outliers los valores por encima del percentil 95 o por debajo del percentil 5.
k = 95
percentile_95 = all_labeled_curado_reducido.quantile(k / 100)
percentile_05 = all_labeled_curado_reducido.quantile((100-k)/100)
```

```python
# visualizamos la ubicación de outliers
ubicacion_outliers = (all_labeled_curado_reducido > percentile_95) | (all_labeled_curado_reducido < percentile_05)
print("\n Ubicacion de outliers \n", ubicacion_outliers)
```

En la matriz precedente figura "True" en caso de que el valor correspondiente esté por encima del 95% o por debajo del 5% (respecto a los datos de la variable correspondiente), figura "False" en caso contrario.

```python
ubicacion_outliers = ubicacion_outliers[~ubicacion_outliers.any(axis=1)]
```

```python
# Acá podemos visualizar la cantidad de registros que en ningún cuantificador tendrían outliers, 
# considerando el percentil 5 (mínimo) y 95 (máximo)
ubicacion_outliers
```

Vemos que son 3420 registros los que no contienen ningún valor atípico (según el criterio seleccionado).

```python
outliers = all_labeled_curado_reducido,[ubicacion_outliers]
print ("\n Lista de outliers\n", outliers)
```

**Habiendo visualizado de manera general los registros que quedarían al aplicar una eliminación de outliers que se encuentren por debajo del percentil 5, ó por encima del percentil 95, nos quedaríamos sólo con el 57% de los datos, por lo que decidimos no eliminarlos ya que estaríamos perdiendo mucha información.<br> 
En el análisis consideramos todos los cuantificadores que cumplan a la vez con dichas condiciones.<br>
Creemos que sería necesario realizar un análisis mas minucioso de los datos atípicos para cada uno de los pacientes ya que en cada una de las métricas podrían tener comportamientos distintos.**

### 3. Balanceo de etiquetas

En el práctico de visualización, registramos que las etiquetas de los datos que distinguen el estado pre-ictal del resto están desbalanceados. Es decir, hay muchos más registros con etiqueta =0 (5536) que los registros con etiqueta =1 (435).

En lo que sigue implementamos algunos métodos para balancear la cantidad de datos con etiqueta 0 o 1, que será nuestro objetivo a predecir en los próximos trabajos. 

- Usamos tres métodos de over-sampling:<br>
a) RandomOverSampler <br>
b) SMOTE <br>
c) ADASYN <br>

*Oversampling* es una técnica que aumentará la cantidad de datos con etiqueta de la que tiene menos registros.

- Usamos tres métodos de under-sampling:<br>
d) RandomUnderSampler<br>
e) ClusterCentroids<br>
f) EditedNearestNeighbours<br>

*Undersampling* es una técnica que disminuirá la cantidad de datos con etiqueta de la que tiene mas registros.

### 5. PCA
